```python
import json

# 1. 模拟：这是从网络爬虫抓回来的一段乱七八糟的 JSON 字符串
raw_api_response = """
{
    "company": "DeepSeek  ",
    "products": [
        {"name": "V2", "price": "free"},
        {"name": "V3", "price": 10},
        {"name": "R1", "price": "unknown"}
    ],
    "status": "success"
}
"""

def parse_market_data(json_str):
    try:
        # 2. 解析 JSON：把字符串变成字典
        data = json.loads(json_str)

        # 3. 清洗数据：去除公司名两边的空格
        company_name = data.get("company", "Unknown").strip()
        print(f"正在分析公司: {company_name}")

        products = data.get("products", [])
        clean_products = []

        # 4. 循环与逻辑判断
        for prod in products:
            price = prod.get("price")

            # 这里的逻辑：如果价格是数字，就存下来；如果是字符串，标记为 -1
            if isinstance(price, (int, float)):
                clean_products.append(prod)
            else:
                print(f"警告: 产品 {prod['name']} 价格数据异常: {price}")

        return clean_products

    except json.JSONDecodeError:
        print("错误: 数据格式不仅是 JSON，无法解析！")
        return []

# 运行函数
result = parse_market_data(raw_api_response)
print("清洗后的数据:", result)
```
